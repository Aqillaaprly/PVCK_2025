```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://i.pinimg.com/736x/99/49/35/99493515f84362378ccb14930c79fac8.jpg",
        "https://i.pinimg.com/736x/cd/4a/80/cd4a804ca63f3a002237d3f3acd6a9d3.jpg",
        "https://i.pinimg.com/736x/26/c7/35/26c7355fe46f62d84579857c6f8c4ea5.jpg"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                       # read image
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # resize image to half size
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB
    final_frame = cv.hconcat([image, image_2])       # concatenate image
    cv2_imshow(final_frame)                          # show image
    print('\n')
```

```python
height = image_2.shape[0]
width = image_2.shape[1]
print("resolusi image: height x width", height, "x",width)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#garis horizontal
for y in range (width):
  image_3[int((height)/2),y] = [255,255,255]
final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Garis vertikal (tengah gambar)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for x in range(height):
    image_3[x, int(width/2)] = [255, 255, 255]

#Garis diagonal (kiri atas ke kanan bawah)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(min(height, width)):
    image_4[i, i] = [255, 255, 255]
for i in range(min(height, width)):
    image_4[height-1-i, i] = [255, 255, 255]

row1 = cv.hconcat([image_2, image_3])
row2 = cv.hconcat([image_2, image_4])
final_frame = cv.vconcat([row1, row2])

cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
y_center = int(height/2)
line_length = 100
x_start = int((width - line_length)/2)
x_end = x_start + line_length
for x in range(x_start, x_end):
    image_3[y_center, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

box_size = 100
x_center = width // 2
y_center = height // 2
x1 = x_center - box_size // 2
y1 = y_center - box_size // 2
x2 = x_center + box_size // 2
y2 = y_center + box_size // 2

# kotak
for y in range(y1, y2):
    for x in range(x1, x2):
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```
